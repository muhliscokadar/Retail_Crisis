```python
import csv
import pandas as pd
import glob
list_1 = []
list_2 = []

files = ["Output/2015_cleandata.csv","Output/2014_cleandata.csv","Output/2013_cleandata.csv","Output/2012_cleandata.csv"]
files2 = ["Output/2011_cleandata.csv","Output/2010_cleandata.csv","Output/2009_cleandata.csv","Output/2008_cleandata.csv",
         "Output/2007_cleandata.csv"]

for i,file in enumerate(files):
    data = pd.read_csv(file)
    del data['Unnamed: 0']
    if i == 0:
        list_1.append(data)
    else:
        del data['description']
        list_1.append(data)
        
    
for i,file in enumerate(files2):
    data = pd.read_csv(file)
    if i == 0:
        list_2.append(data)
    else:
        del data['NAICS Description']
        list_2.append(data)
type(list_2[0])
```

```python
full_df = pd.merge(list_1[0],list_1[1],on='code',suffixes = ('_2015','_2014'))
full_df = pd.merge(full_df,list_1[2],on='code',suffixes = ('','_2013'))
full_df = pd.merge(full_df,list_1[3],on='code',suffixes = ('','_2012'))
full_df
```

```python
other_df = pd.merge(list_2[0],list_2[1],on='NAICS Code',suffixes = ('_2011','_2010'))
other_df = pd.merge(other_df,list_2[2],on='NAICS Code',suffixes = ('','_2009'))
other_df = pd.merge(other_df,list_2[3],on='NAICS Code',suffixes = ('','_2008'))
other_df = pd.merge(other_df,list_2[4],on='NAICS Code',suffixes = ('','_2007'))
del other_df['Unnamed: 0']
other_df = other_df.rename(columns = {'code':'NAICS Code',
                                    'Number of Establishments':'Number of Establishments_2009',
                                    'Employees':'Employees_2009'})

other_df
```

```python
full_df = full_df.rename(columns = {'code':'NAICS Code',
                                    'establishments':'establishments_2013',
                                    'employment':'employment_2013'})
full_df
```

```python
complete_df = pd.merge(full_df,other_df,on="NAICS Code")
complete_df = complete_df.rename(columns = {'Number of Establishments_2011':'establishments_2011',
                                            'Employees_2011':'employment_2011',
                                           'Number of Establishments_2010':'establishments_2010',
                                            'Employees_2010':'employment_2010',
                                           'Number of Establishments_2009':'establishments_2009',
                                            'Employees_2009':'employment_2009',
                                           'Number of Establishments_2008':'establishments_2008',
                                            'Employees_2008':'employment_2008',
                                           'Number of Establishments_2007':'establishments_2007',
                                            'Employees_2007':'employment_2007',})
complete_df
```

```python
del complete_df['NAICS Description']
```

```python
complete_df.to_csv("Output/merged_0715_data.csv", index = False)
```
