```python
import os
import csv
import pandas as pd
```

```python
datafile = ( "../raw_data/retaildata_2001.csv")
data_zero = pd.read_csv(datafile)
```

```python
df=pd.DataFrame(data_zero)
df
```

```python
df2001 = df.rename(columns=df.iloc[0]).drop(df.index[0])
```

```python
df2001
```

```python
df2001=df2001.loc[df2001['CODE'].isin(["442", "443","4451","448"])]
```

```python
df2001
```

```python
#locates specific data from the Data Type and Code columns


t=df2001.loc[(df2001['DATA TYPE'] == "Establishments")|(df2001['DATA TYPE'] == "Employment") & (df2001['CODE'].isin(["442","443","4451","448"]))]
t
```

```python
# only pulls in the first two columns from t dafaframe.

newdf2001 = t[t.columns[:2]]
newdf2001
```

```python
#makes Data Type column into seperate columns
testing = t.pivot(index="CODE",columns='DATA TYPE', values='TOTAL')
testing
```

```python
#merge the two dataframes
mergedf=pd.merge(newdf2001, testing, how='left', on='CODE')
mergedf
```

```python
#drops duplicate rows to make a final datatframe
final=mergedf.drop_duplicates(subset=['CODE'])
final
```

```python
final2001=final.rename(columns={'Employment': 'Employment_2001','Establishments':'Establishments_2001'})
final2001
```

```python
#save the final datatframe as a csvfile
final2001.to_csv("Output/2001_cleandata.csv",header=True,index=False)
```
