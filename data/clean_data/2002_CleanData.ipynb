```python
####Columns need to be renamed or dropped depending on how we decide to use the information.


import os
import csv
import pandas as pd
```

```python
datafile = ( "../raw_data/retaildata_2002.csv")
two_thousand_two = pd.read_csv(datafile)
```

```python
df=pd.DataFrame(two_thousand_two)
df
```

```python
df.columns
```

```python
df=df.rename(columns={"Employment Size of the Enterprise": "0", "Unnamed: 5":"1-4", "Unnamed: 6":"5-9",
       "Unnamed: 7":"10-19", "Unnamed: 8":"<20", "Unnamed: 9":"20-99", "Unnamed: 10":"100-499", "Unnamed: 11":"<500",
       "Unnamed: 12":"500+"})
```

```python
df
```

```python
df2002 = df.drop(df.index[0])
```

```python
df2002
```

```python
df2002=df2002.loc[df2002['NAICS CODE'].isin(["442", "443","4451","448"])]
```

```python
df2002
```

```python
len(df2002)
```

```python
t=df2002.loc[(df2002['DATA TYPE'] == "Establishments")|(df2002['DATA TYPE'] == "Employment") & (df2002['NAICS CODE'].isin(["442","443","4451","448"]))]
t
```

```python
newdf2002 = t[t.columns[:2]]
newdf2002
```

```python
 #makes Data Type column into seperate columns
testing = t.pivot(index="NAICS CODE",columns='DATA TYPE', values='TOTAL')
testing
```

```python
mergedf=pd.merge(newdf2002, testing, how='left', on='NAICS CODE')
final=mergedf.drop_duplicates(subset=['NAICS CODE'])
#mergedf.rename(columns={'NAICS CODE': "CODE", 'Employment': 'Employment_2002','Establishments':'Establishments_2002'})
```

```python
final
```

```python
final2002=final.rename(columns={'NAICS CODE': "CODE", 'Employment': 'Employment_2002','Establishments':'Establishments_2002'})
final2002
```

```python
#save the final datatframe as a csvfile
final2002.to_csv("Output/2002_cleandata.csv",header=True,index=False)
```
