```python
####Columns need to be renamed or dropped depending on how we decide to use the information.


import os
import csv
import pandas as pd
```

```python
datafile = ( "../raw_data/retaildata_2004.csv")
two_thousand_four = pd.read_csv(datafile)
```

```python
df=pd.DataFrame(two_thousand_four)
df
```

```python
df2004 = df.rename(columns=df.iloc[1]).drop(df.index[0])
df2004
```

```python
df4=df2004.drop(df2004.index[0])
df4
```

```python
df4=df4.loc[df4['CODE'].isin(["442", "443","4451","448"])]
```

```python
df4
```

```python
t=df4.loc[(df4['DATA TYPE'] == "Establishments")|(df4['DATA TYPE'] == "Employment") & (df4['CODE'].isin(["442","443","4451","448"]))]
t
```

```python
newdf2004 = t[t.columns[:2]]
newdf2004
```

```python
 #makes Data Type column into seperate columns
testing = t.pivot(index="CODE",columns='DATA TYPE', values='TOTAL')
testing
```

```python
mergedf=pd.merge(newdf2004, testing, how='left', on='CODE')
final=mergedf.drop_duplicates(subset=['CODE'])
#mergedf.rename(columns={'NAICS CODE': "CODE", 'Employment': 'Employment_2002','Establishments':'Establishments_2002'})
```

```python
final
```

```python
final2004=final.rename(columns={'Employment': 'Employment_2004','Establishments':'Establishments_2004'})
final2004
```

```python
#save the final datatframe as a csvfile
final2004.to_csv("Output/2004_cleandata.csv",header=True,index=False)
```
