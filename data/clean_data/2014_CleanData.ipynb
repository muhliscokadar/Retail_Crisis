```python
import os 
import csv 
import pandas as pd
```

```python
datapath = ("../raw_data/retaildata_2014.csv")
datafile = pd.read_csv(datapath, encoding="ISO-8859-1")
```

```python
df=pd.DataFrame(datafile)
df
```

```python
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
```

```python
df.columns
```

```python
df2014= df.loc[df[' NAICS   CODE '].isin(["442", "443","4451","448"])]
df2014
```

```python
df['ENTERPRISE EMPLOYMENT SIZE']
```

```python
df2014= df2014.loc[df2014['ENTERPRISE EMPLOYMENT SIZE'].isin(["01:  Total"])]
df2014
```

```python
clean2014 = df2014.drop(columns=['ENTERPRISE EMPLOYMENT SIZE', 'NUMBER OF FIRMS', 'EMPLOYMENT NOISE FLAG',
                                'ANNUAL PAYROLL NOISE FLAG', 'ANNUAL PAYROLL ($1,000)', 'EMPLOYMENT RANGE FLAG'
                                ])
clean2014
```

```python
clean2014.columns
```

```python
clean2014=clean2014.rename(columns={' NAICS   CODE ':'code',
                                   'NAICS DESCRIPTION': 'description',
                                   'NUMBER OF ESTABLISHMENTS':'establishments',
                                    'EMPLOYMENT':'employment'})
clean2014
```

```python
clean2014.to_csv("Output/clean_2014retail.csv", index=True, header=True)
```
