```python
####Columns need to be renamed or dropped depending on how we decide to use the information.


import os
import csv
import pandas as pd
```

```python
datafile = ( "../raw_data/retaildata_2003.csv")
two_thousand_three = pd.read_csv(datafile)
```

```python
df=pd.DataFrame(two_thousand_three)
df
```

```python
df2003 = df.rename(columns=df.iloc[1]).drop(df.index[0])
df2003
```

```python
df3=df2003.drop(df2003.index[0])
df3
```

```python
df3=df3.loc[df3['CODE'].isin(["442", "443","4451","448"])]
```

```python
df3
```

```python
t=df3.loc[(df3['DATA TYPE'] == "Establishments")|(df3['DATA TYPE'] == "Employment") & (df3['CODE'].isin(["442","443","4451","448"]))]
t
```

```python
newdf2003 = t[t.columns[:2]]
newdf2003
```

```python
 #makes Data Type column into seperate columns
testing = t.pivot(index="CODE",columns='DATA TYPE', values='TOTAL')
testing
```

```python
mergedf=pd.merge(newdf2003, testing, how='left', on='CODE')
final=mergedf.drop_duplicates(subset=['CODE'])
#mergedf.rename(columns={'NAICS CODE': "CODE", 'Employment': 'Employment_2002','Establishments':'Establishments_2002'})
```

```python
final
```

```python
final2003=final.rename(columns={'Employment': 'Employment_2003','Establishments':'Establishments_2003'})
final2003
```

```python
#save the final datatframe as a csvfile
final2003.to_csv("Output/2003_cleandata.csv",header=True,index=False)
```
