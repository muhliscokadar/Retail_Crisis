```python
import os 
import csv 
import pandas as pd
```

```python
datapath = ("../raw_data/retaildata_2015.csv")
datafile = pd.read_csv(datapath, encoding="ISO-8859-1")
```

```python
df=pd.DataFrame(datafile)
df
```

```python
df.columns
```

```python
df2015= df.loc[df[' NAICS   CODE '].isin(["442", "443","4451","448"])]
df2015
```

```python
df['ENTERPRISE EMPLOYMENT SIZE']
```

```python
df2015= df2015.loc[df2015['ENTERPRISE EMPLOYMENT SIZE'].isin(["01:  Total"])]
df2015
```

```python
clean2015 = df2015.drop(columns=['ENTERPRISE EMPLOYMENT SIZE', 'NUMBER OF FIRMS', 'EMPLOYMENT NOISE FLAG',
                                'ANNUAL PAYROLL NOISE FLAG', 'ANNUAL PAYROLL ($1,000)', 'EMPLOYMENT RANGE FLAG'
                                ])
clean2015
```

```python
clean2015.columns
```

```python
clean2015=clean2015.rename(columns={' NAICS   CODE ':'code',
                                   'NAICS DESCRIPTION': 'description',
                                   'NUMBER OF ESTABLISHMENTS':'establishments',
                                    'EMPLOYMENT':'employment'})
clean2015
```

```python
clean2015.to_csv("Output/clean_2015retail.csv", index=True, header=True)
```
